```python
#Importando os pacotes das bibliotecas
import pandas as pd
import numpy as np
```

```python
#Carregando a base de dados
data = pd.read_pickle("base_submission.pkl")
data.head()
```

```python
#Obter as dimensões do dataframe
data.shape
```

```python
#Imprime informações sobre um DataFrame, incluindo o tipo de índice e colunas, valores não nulos e uso de memória.
data.info()
```

```python
#Divisão entre previsores e classes
X_data = data.iloc[:, 7:13]
X_data.head()
```

```python
import joblib
modelo = joblib.load("modelo.pkl")
```

```python
#Realizando as previsões com o modelo treinado
previsoes = modelo.predict(X_data)

#Salvando o arquivoo 
joblib.dump(previsoes, "resultado.pkl")
print("Previsões Geradas com Sucesso!")
```
